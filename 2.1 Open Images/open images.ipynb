# Чтение, запись изображений. Работа с пикселями

```python
%matplotlib inline
# Служебная команда, позволяющая просматривать изображения в браузере
```

```python
from skimage.io import imread, imshow, imsave
```

```python
img = imread('tiger.jpg')
```

```python
imshow(img)
```

```python
img.shape
# Размеры изображения. Функция выводит (i, j, k)
# i = кол-во строк
# j = кол-во стобцов
# k = кол-во каналов
```

```python
img.shape[0]
# 0 - строки 
# 1 - стобцы
# 2 - каналы
```

```python
# координаты пикселя
img[525, 525]
# вывод: ([R, G, B], тип)
```

```python
# цвет канала вход [x, y, канал]
img[525, 525, 1]
# вывод: насыщенность цвета 0-255
# 0 - R
# 1 - G
# 2 - B
```

```python
img[525, 525] = [255, 255, 0]
# смена цвета пикселя
```

```python
# сохраним изображение 
# imsave(путь к сохр файлу, переменная с файлом)
imsave('tiger-new.jpg', img)
```
